```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25
#The +ve values on the kernel allows filtering brightness, while -ve highlights the darkness and 0 the grey from the original image when the filter slides on top of input image
#
kernel = np.float32([[-1,0,1],[-2,0,2],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

**Horizontal Edge Detector**

To detect the  horizontal edge we need to increase the pixels of horizontal edges and reduce the pixels of other edges, the pixels in the horizontal edges should be mulitplied with possitive kernal values and rest of the edges should be multiplied with negetive kernal values.

**NOTE**:Kernal values user based on traila and error but in deeep learning network will learn valsue through backpropgation


```
#kernel = np.ones((3,3),np.float32)/25
# Horizontal Edge Detector
kernel = np.float32([[1,2,1],[0,0,0],[-1,-2,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

**45 Degree Angle Detector**

to detect angled edge, the pixels of angled edges should be increased and the rest of the pixels should be decreased .to get that we use a kernal with higher values diagonally.

**NOTE**:Kernal values user based on traila and error but in deeep learning network will learn valsue through backpropgation

```
# 45 Degree Angle Detector
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[2,-1,-1],[-1,2,-1],[-1,-1,2]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

**Blur Kernel**
To blur image all kernal values should be normalizd with lower values


**NOTE**:Kernal values user based on traila and error but in deeep learning network will learn valsue through backpropgation

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[0.0625,-0.125,0.0625],[0.125,-0.25,0.125],[0.0625,-0.125,0.0625]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
# Sharpen Kernel
#kernel = np.ones((3,3),np.float32)/25
# kernel = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])
kernel = np.float32([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
# Identity function

kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])
dst = cv2.filter2D(image,-1, kernel)
cv2_imshow(dst)
```
